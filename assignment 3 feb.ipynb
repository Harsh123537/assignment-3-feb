```python
# *args is used to passed as much as argument as required to the function. It returns the tuple
# **kwargs is used to return the dictionary which were passed as argument to function

def func1(*var):
    return var #return tuple
print(func1(1,2,3,4,5,6,"fkksdcjsdc",[3,4,5,3,3,2,1]))

def func2(**var):
    return var #return dictionary
print(func2(a=4,b="rfdfds",c=[3,6,7,8,9]))

```

```python
# generator fnction is used to print the value through the yield keyword without storing all the data
# in the memory.
# yield keyword is used to print the value for particular iteration.

def func(n):
    for i in range(n):
        yield i
a=func(10)
for i in a:
    print(i)
    
```

```python
def prime(n):
    cnt=0
    for i in range(2,n):
        for j in range(1,(i//2)+1):
            if i%j==0:
                cnt=cnt+1
        if cnt==1:
            yield i
        cnt=0
a=prime(1000)
for i in range(20):
    print(next(a))
        
            
```
